```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score 
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
for column in df.columns:
        if len(df[column].value_counts()) == 1:
            print(str(column) + ": " + str(df[column].unique()))
            print(df[column].value_counts())
            print("===================================================================")
```

```python
def drop(dfs):
    dfs.drop(columns = ["id","EmployeeCount","Over18","StandardHours"], axis =1, inplace = True)

drop(df)
```

```python
df.head()
```

```python
for column in df.columns:
    if df[column].dtype == object:
        print(str(column) + ": " + str(df[column].unique()))
        print(df[column].value_counts())
        print("===================================================================")
```

```python
def Encoder(dfs):
    le = LabelEncoder()
    for column in dfs.columns:
        if dfs[column].dtype == np.number:
            continue
        else:
            dfs[column] = le.fit_transform(dfs[[column]])

Encoder(df)
```

```python
def Scaler(dfs):
    scaler = MinMaxScaler()
    for column in dfs.columns:
        dfs[column] = scaler.fit_transform(dfs[[column]])

Scaler(df)
```

```python
df.head()
```

```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.metrics import roc_auc_score

model = RandomForestClassifier(n_estimators = 3000, criterion = "gini", min_samples_leaf = 1, max_features ="sqrt", warm_start = True)
skf = StratifiedKFold(n_splits = 80, shuffle = True, random_state = 60)

best_fold = None
best_classification_report = None
best_mean_roc_auc = 0.0

X = df.drop("Attrition", axis = 1)
y = df.Attrition

New_df = pd.read_csv("test.csv")
drop(New_df)
Encoder(New_df)
Scaler(New_df)

for fold,  (train_index, test_index) in enumerate(skf.split(X,y)):
    x_train, x_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    
    model.fit(x_train,y_train)
    
    y_pred = model.predict(x_test)
    
#     y_proba = model.predict_proba(x_test)[:,1]
    New_predict = model.predict_proba(New_df)[:,1]
    
    model_rep = classification_report(y_test, y_pred)
    
    roc_score = roc_auc_score(y_test, y_pred)
    
    accuracy = accuracy_score(y_test, y_pred)
    
#     print(f'Roc Score : {roc_score}\n')
#     print(f"Fold {fold + 1} Classification Report:\n{model_rep}")
    
#     updating report if current fold have higher roc_auc_score
    if roc_score > best_mean_roc_auc:
        best_mean_roc_auc = roc_score
        best_fold = fold + 1
        best_classification_report = model_rep
        # predicting attrition for the new test data
#         New_predict = model.predict_proba(New_df)[:,1]

# printing the best classification report, roc_score, and best fold
print(f"Accuracy Score {accuracy:.4f}\n")
print(f"Best Fold {best_fold}\n")
print(f"{best_classification_report}")
print(f"ROC AUC : {best_mean_roc_auc:.4f}")

# new_data_indices = next(skf.split(New_df))[1]
```
